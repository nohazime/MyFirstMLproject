```python
#histogram
def hist(x):
    plt.hist(df[x], bins=25)
    plt.title(x, fontsize=10, loc="right")
    plt.xlabel('Relative frequency')
    plt.ylabel('Absolute frequency')
    plt.show()

# General statistics
def stats(x):
    print(f"Variable: {x}")
    print(f"Type of variable: {df[x].dtype}")
    print(f"Total observations: {df[x].shape[0]}")
    detect_null_val = df[x].isnull().values.any()
    if detect_null_val:
        print(f"Missing values: {df[x].isnull().sum()} ({(df[x].isnull().sum() / df[x].isnull().shape[0] *100).round(2)}%)")
    else:
        print(f"Missing values? {df[x].isnull().values.any()}")
        print(f"Unique values: {df[x].nunique()}")
    if df[x].dtype != "O":
        print(f"Min: {int(df[x].min())}")
        print(f"25%: {int(df[x].quantile(q=[.25]).iloc[-1])}")
        print(f"Median: {int(df[x].median())}")
        print(f"75%: {int(df[x].quantile(q=[.75]).iloc[-1])}")
        print(f"Max: {int(df[x].max())}")
        print(f"Mean: {df[x].mean()}")
        print(f"Std dev: {df[x].std()}")
        print(f"Variance: {df[x].var()}")
        print(f"Skewness: {scipy.stats.skew(df[x])}")
        print(f"Kurtosis: {scipy.stats.kurtosis(df[x])}")
        print("")

        # Percentiles 1%, 5%, 95% and 99%
        print("Percentiles 1%, 5%, 95%, 99%")
        display(df[x].quantile(q=[.01, .05, .95, .99]))
        print("")
    else:
        print(f"List of unique values: {df[x].unique()}")

# Bar chart
def bar(x):
    ax = df[x].value_counts().plot(kind="bar", figsize=(6,5), fontsize=10, color=sns.color_palette("rocket"), table=False)
    for p in ax.patches:
        ax.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')
    plt.xlabel(x, fontsize=10)
    plt.xticks(rotation=0, horizontalalignment="center")
    plt.ylabel("Absolute values", fontsize=10)
    plt.title(x, fontsize=10, loc="right")

# Pivot_table_sum
def pivot_sum(a, b, c):
    type_pivot_sum = df.pivot_table(
        columns=a,
        index=b,
        values=c, aggfunc=np.sum)
    display(type_pivot_sum)
    type_pivot_sum.sort_values(by=[b], ascending=True).plot(kind="bar", title=(b), figsize=(6,4),fontsize = 12);

# Variable vs. target chart
def target(x):
    short_0 = df[df.bad_loan == 0].loc[:,x]
    short_1 = df[df.bad_loan == 1].loc[:,x]

    a = np.array(short_0)
    b = np.array(short_1)

    np.warnings.filterwarnings('ignore')

    plt.hist(a, bins=40, density=True, color="g", alpha = 0.6, label='Not-default', align="left")
    plt.hist(b, bins=40, density=True, color="r", alpha = 0.6, label='Default', align="right")
    plt.legend(loc='upper right')
    plt.title(x, fontsize=10, loc="right")
    plt.xlabel('Relative frequency')
    plt.ylabel('Absolute frequency')
    plt.show()

def boxhist(x):
    variable = df[x]
    np.array(variable).mean()
    np.median(variable)
    f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw= {"height_ratios": (0.5, 2)})
    mean=np.array(variable).mean()
    median=np.median(variable)
    sns.boxplot(variable, ax=ax_box)
    ax_box.axvline(mean, color='r', linestyle='--')
    ax_box.axvline(median, color='g', linestyle='-')
    sns.distplot(variable, ax=ax_hist)
    ax_hist.axvline(mean, color='r', linestyle='--')
    ax_hist.axvline(median, color='g', linestyle='-')
    plt.title(x, fontsize=10, loc="right")
    plt.legend({'Mean':mean,'Median':median})
    ax_box.set(xlabel='')
    plt.show()

# Pivot_table_mean
def pivot_mean(a, b, c):
    type_pivot_mean = df.pivot_table(
        columns=a,
        index=b,
        values=c, aggfunc=np.mean)
    display(type_pivot_mean)
    type_pivot_mean.sort_values(by=[b], ascending=True).plot(kind="bar", title=(b), figsize=(6,4),fontsize = 12);
```

```python
# ROC Curve: Area Under the Curve
def auc_roc_plot(y_test, y_preds):
    fpr, tpr, thresholds = roc_curve(y_test,y_preds)
    roc_auc = auc(fpr, tpr)
    print(roc_auc)
    plt.title('Receiver Operating Characteristic')
    plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
```

```python
def transform_instance(new_instance):
  #filtering numeric and categoric data
  num_features=new_instance.select_dtypes(np.number)
  cat_features=new_instance.select_dtypes(include=[np.object])

  # Grade mapping
  cat_features["grade"] = grade_mapping(cat_features["grade"])

  # One-hot encoding
  OneHot_term = pd.DataFrame(onehot_term.transform(cat_features[['term']]).toarray(), columns=onehot_term.get_feature_names_out(input_features=['term']))
  OneHot_homeownership = pd.DataFrame(onehot_home.transform(cat_features[['home_ownership']]).toarray(), columns=onehot_home.get_feature_names_out(input_features=['home_ownership']))

  cat_features = cat_features.join(OneHot_term)
  cat_features = cat_features.join(OneHot_homeownership)

  # Binary encoding
  bin_features= bin_enc_purp.transform(cat_features['purpose'])
  cat_features=pd.concat([cat_features,bin_features],axis=1)
  cat_features.drop(["purpose","term","home_ownership"], axis=1, inplace=True)

  data=pd.concat([cat_features,num_features],axis=1)
  data.columns = data.columns.astype(str)

  return data.head(1)
```

```python
%%writefile app.py
import streamlit as st
import pickle
import numpy as np
import pandas as pd


def load_model():
  with open("model_lg.pkl", "rb") as file:
    data=pickle.load(file)
  return data

data=load_model()

model_lg=data["model"]
bin_enc_purp=data["purpose"]
onehot_term=data["term"]
onehot_home=data["home_ownership"]
mean=data["train data mean"]
std=data["train data std"]


def classify(num):
  if num>0.5 :
    return 'Refused'
  else:
    return 'Accepted'

def grade_mapping(input_grade):
  grades = {"A": 7, "B": 6, "C": 5, "D": 4, "E": 3, "F": 2, "G": 1}
  input_grade = [grades.get(val, 0) for val in input_grade]  # Utilisez get avec une valeur par défaut de 0
  return input_grade

def transform_instance(new_instance):
  #filtering numeric and categoric data
  num_features=new_instance.select_dtypes(np.number)
  cat_features=new_instance.select_dtypes(include=[np.object])
  # Grade mapping
  cat_features["grade"] = grade_mapping(cat_features["grade"])
  # One-hot encoding
  OneHot_term = pd.DataFrame(onehot_term.transform(cat_features[['term']]).toarray(), columns=onehot_term.get_feature_names_out(input_features=['term']))
  # Encodage one-hot pour la colonne 'home_ownership'
  OneHot_homeownership = pd.DataFrame(onehot_home.transform(cat_features[['home_ownership']]).toarray(), columns=onehot_home.get_feature_names_out(input_features=['home_ownership']))

  enc_data =OneHot_term.join(OneHot_homeownership, lsuffix="_term", rsuffix="_home")
  cat_features = cat_features.join(enc_data)

  # Binary encoding
  bin_features= bin_enc_purp.transform(cat_features['purpose'])
  cat_features=pd.concat([cat_features,bin_features],axis=1)
  cat_features.drop(["purpose","term","home_ownership"], axis=1, inplace=True)

  data=pd.concat([cat_features,num_features],axis=1)
  data.columns = data.columns.astype(str)

  return data.head(1)



def show_predict_page():
    st.title("Predicting Bank Loan Defaults")

    st.write("""### We need some information to predict the loan""")

    purposes = (
        "debt_consolidation",
        "credit_card",
        "other",
        "home_improvement",
        "major_purchase",
        "small_business",
        "car",
        "medical",
        "moving",
        "vacation",
        "house",
        "wedding",
    )

    homeownerships = (
        "RENT",
        "OWN",
        "MORTGAGE",
    )

    terms = (
        " 36 months" ,
        " 60 months",
    )

    grades = (
        "A",
        "B",
        "C",
        "D",
        "E",
        "F",
        "G",
    )

    purpose = st.selectbox("purpose", purposes)
    grade = st.selectbox("grade", grades)
    homeownership = st.selectbox("home ownership", homeownerships)
    term = st.selectbox("term", terms)

    annual_inc = st.number_input("Annual Income", value=0)
    short_emp = st.selectbox("Short Employment",[0,1])
    emp_length_num = st.slider("Employment length in years", 0, 10, 5)
    dti=st.number_input("Debt-To-Income Ratio", value=0)
    last_delinq_none =st.selectbox("1 when the borrower had at least one event of delinquency",[0,1])
    revol_util = st.number_input("Revolving line utilization rate", value=0)
    total_rec_late_fee=st.number_input("Late fees received to date", value=0)

    ok = st.button("predict loan")
    if ok:
        X_ = np.array([['grade', 'homeownership', 'purpose', 'term', annual_inc, short_emp,emp_length_num, dti, last_delinq_none, revol_util, total_rec_late_fee]])
        df_x = pd.DataFrame(X_, columns=['grade', 'home_ownership', 'purpose', 'term', 'annual_inc', 'short_emp','emp_length_num', 'dti', 'last_delinq_none', 'revol_util', 'total_rec_late_fee'])
        # Convert numerical columns to appropriate data types
        df_x['annual_inc'] = df_x['annual_inc'].astype(float)
        df_x['short_emp'] = df_x['short_emp'].astype(int)
        df_x['emp_length_num'] = df_x['emp_length_num'].astype(int)
        df_x['dti'] = df_x['dti'].astype(float)
        df_x['last_delinq_none'] = df_x['last_delinq_none'].astype(int)
        df_x['revol_util'] = df_x['revol_util'].astype(float)
        df_x['total_rec_late_fee'] = df_x['total_rec_late_fee'].astype(float)
        input_data_encoded =  transform_instance(df_x)
        input_data_encoded=(input_data_encoded-mean)/std
        loann = model_lg.predict(input_data_encoded)
        result=classify(loann)
        st.subheader(f"The loan is {result}")


show_predict_page()
```
